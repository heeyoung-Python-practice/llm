```python
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper

wiki_tool = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper())
wiki_tool.run("í”¼ì§€ì»¬ AI")
```

```python
from langchain.chat_models import init_chat_model
from langchain.agents import create_agent   # Tool ì‚¬ìš© ì—ì´ì „íŠ¸ ìƒì„±
from pprint import pprint   # ë³´ê¸° ì¢‹ê²Œ ì¶œë ¥

messages = [
    ('human', 'ì˜¬ë°ì´ í”„ë¡œì íŠ¸ ë©¤ë²„ ì•Œë ¤ì¤˜!')
]

llm = init_chat_model('openai:gpt-4.1-mini')
# print(llm.invoke(message))

agent = create_agent(
    model = llm,
    tools = [wiki_tool]
)

response = agent.invoke({'messages': messages}) # messagesë¥¼ dicë¡œ ë„£ì–´ ì—ì´ì „í…Œã…¡ ì‹¤í–‰

print(response)
```
